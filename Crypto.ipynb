```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df= pd.read_csv("ETH-BTC-USD.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
sns.countplot(data=df, x=df['Currency'])
```

```python
sns.displot(data=df, x='High', col='Currency',kind='kde')
```

```python
df['NuevaDate']=(df['Date']).str[-4:]
df.head()
```

```python
sns.countplot(data=df, x=df['NuevaDate'])
```

```python
df_2018=df[df['NuevaDate']=='2018']
df_2018
```

```python
df_2018.index
```

```python
df_2018=pd.DataFrame(df_2018)
```

```python
df_2018.shape[0]
```

```python
#for i in range(df_2018.shape[0]):
 # print(i)

#aqu ya arreglamos los indices
df_2018.index=range(df_2018.shape[0])
df_2018
```

```python
sns.barplot(data=df_2018, x="Date", y='High')
```

```python
df_2018['Date'] = pd.to_datetime(df_2018['Date'], format='%m/%d/%Y')
sns.scatterplot(data=df_2018,x='Date', y='High')
plt.xlabel('Fecha')
plt.ylabel('Valores de la crypto')
plt.show()
```

```python
sns.lineplot(data=df_2018,x='Date', y='High')
plt.xlabel('Fecha')
plt.ylabel('Valores de la crypto')
plt.show()
```
