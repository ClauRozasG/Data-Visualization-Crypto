```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df= pd.read_csv("ETH-BTC-USD.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
sns.countplot(data=df, x=df['Currency'])
```

```python
sns.displot(data=df, x=df['Date'].dt.year, y='High', col='Currency',kind='kde')
```

```python
df['NuevaDate']=(df['Date']).str[-4:]
df.head()
```

```python
sns.countplot(data=df, x=df['NuevaDate'])
```

```python
df_2018=df[df['NuevaDate']=='2018']
df_2018
```

```python
area=plt.figure()
g4=area.add_subplot(2,2,1)
```

```python
grupo=df.groupby('NuevaDate')['Open']
label=grupo.mean().index
valores=grupo.mean().values
g4.bar(label,valores)
g4.set_title('Valor promedio de valor de apertura')
area
```

```python
grupo = df.groupby(['NuevaDate', 'Currency'])['Open'].mean().reset_index()
sns.barplot(x='NuevaDate', y='Open', hue='Currency', data=grupo)
plt.xlabel('Fecha')
plt.ylabel('Valor medio de Open')
plt.title('Media de Open por fecha y moneda')
plt.show()
```

```python
grupo = df.groupby(['NuevaDate', 'Currency'])['High'].mean().reset_index()
sns.barplot(x='NuevaDate', y='High', hue='Currency', data=grupo)
plt.xlabel('Fecha')
plt.ylabel('Valor medio de High')
plt.title('Media de High por fecha y moneda')
plt.show()
```

```python
grupo = df.groupby(['NuevaDate', 'Currency'])['Low'].mean().reset_index()
sns.barplot(x='NuevaDate', y='Low', hue='Currency', data=grupo)
plt.xlabel('Fecha')
plt.ylabel('Valor medio de Low')
plt.title('Media de Low por fecha y moneda')


plt.show()
```
